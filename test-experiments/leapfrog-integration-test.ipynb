```python
import tensorflow as tf
import matplotlib.pyplot as plt

# use ggplot styles for graphs
plt.style.use('ggplot')

from bayes_vi.utils.leapfrog_integrator import LeapfrogIntegrator
```

# Leapfrog Integration Test

```python
initial_state = tf.constant([[0.,1.]])
leapfrog = LeapfrogIntegrator()
kinetic_energy = lambda p: 0.5 * p**2
potential_energy = lambda q: 0.5 * q**2
```

```python
path = [initial_state]
for _ in range(50):
    path.append(leapfrog.solve(potential_energy, kinetic_energy, path[-1], 0.1, 10))
```

```python
path_tensor = tf.stack(tf.squeeze(path), axis=0)
```

```python
plt.scatter(*tf.split(tf.squeeze(path_tensor), num_or_size_splits=2, axis=1))
```
